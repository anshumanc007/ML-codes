```

def initialise(dim):
  W=np.random.rand(dim,1)
  b=0.00
  return W,b
```

```
def propagate(W,b,X,Y):
  m=X.shape[1]
  A=(np.dot(W.T,X)+b);
  cost=1/(2*m)*np.sum((A-Y)**2);
  dW=1/m*np.dot(X,(A-Y).T);
  db=1/m*np.sum(A-Y);
  grads = {"dW": dW,
             "db": db};
  return grads,cost;
    
```

```
def optimise(W,b,X,Y,num_iteration,alpha):
  costs=[]
  for i in range(num_iteration):
    grads,cost=propagate(W,b,X,Y)
    dW=grads['dW'];
    db=grads['db'];
    W=W-alpha*dW;
    b=b-alpha*db;
    costs.append(cost);
    #if (i%100==0):
     # print(cost)
    
  parameters={"W": W,
                "b":b
              
               }
  gradients={"dW":dW,
               "db":db}
   
  return parameters,gradients
```

```
def prediction(W,b,X):
  m=X.shape[1];
  A=np.dot(W.T,X)+b;     
  return A;      
    
```

```
def model(X_train,Y_train,X_test,Y_test,num_iteration,alpha):
  mini=100000;
  for i in range(100):
    W,b=initialise(X_train.shape[0])
    parameters, grads = optimise(W, b, X_train, Y_train, num_iteration,alpha)
    W=parameters['W'];
    b=parameters['b'];
    Y_prediction_test = prediction(W, b, X_test)
    Y_prediction_train = prediction(W,b,X_train)
    m=X_test.shape[1]
    error=1/m*np.sum((Y_prediction_test-Y_test)**2)
    if(error<mini):
      mini=error
  
    
  d = {  "Y_prediction_test": Y_prediction_test, 
         "Y_prediction_train" : Y_prediction_train, 
          "error" : mini,
         "W" : W, 
         "b" : b,
         "learning_rate" : alpha,
         "num_iterations": num_iteration};
  return d;
 
  
  
```
